```python
monsters = pd.read_csv("monsterManual5e.csv")

monsters.loc[monsters['type'] == ' monstrosity', 'type'] = 'monstrosity'
```

```python
def get_subgroup(x): 
    sub_group = x[x.find("(")+1:x.find(")")]
    sub_group = re.sub('(|)', '', sub_group)
    return sub_group

def get_supergroup(x): 
    super_group = re.sub('[\(].*?[\)]', '', x).rstrip()
    return super_group

monsters['sub_type'] = monsters['type'].apply(get_subgroup)
monsters['type'] = monsters['type'].apply(get_supergroup)
```
